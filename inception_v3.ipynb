```python
import torch
import torchvision.models as models

# Load the InceptionV3 model
model = models.inception_v3(pretrained=True)
model.eval()

# Accessing the desired layer
desired_layer_name = "Block1_Conv1"  # Example layer name
desired_layer = None

# Iterate through model's named children to find the desired layer
for name, layer in model.named_children():
    print(name)
    if desired_layer_name in name:
        desired_layer = layer
        break

if desired_layer is None:
    print("Desired layer not found")
else:
    print(desired_layer)
```
